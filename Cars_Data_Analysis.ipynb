##**Project set-up**


> Import pandas

```python
import pandas as pd
```

> Use the raw dataset to load a datatframe

```python
# Example URL, replace with your actual raw URL
url = 'https://raw.githubusercontent.com/ultrasage-danz/Cars-Data-Analysis/main/Project%2B2%2B-%2BCars%2BDataset.csv'
df = pd.read_csv(url)
df.head()
# df.shape
```

##**Data Cleaning**



> Find all null values in the dataset



```python
df.isnull().sum()
```

> Replace all null values with mean of column

```python
df['Cylinders'].fillna(df['Cylinders'].mean(), inplace=True)
```

##**Value-count checks**

> Different Car makes in the data.

```python
df['Make'].value_counts()
```

>  Occurrence of each car make in the data - using a bar chart

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
plt.figure(figsize=(10,5))

# Generate a colormap with 20 colors
num_colors = 38
colormap = cm.get_cmap('rainbow', num_colors)

# Plot the bar chart with the colormap
df['Make'].value_counts().sort_values().plot(kind='bar', color=colormap(range(num_colors)))


# Add axes and a title
plt.xlabel('Car Make')
plt.ylabel('Count')
plt.title('Car Make Distribution')

# Show the plot
plt.show()
```
